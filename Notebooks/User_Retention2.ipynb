# Model Development

```python
X_train = train_df[['Video position (%)']]
y_train = train_df['Absolute audience retention (%)']

# Train a regression model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
video_A = pd.read_excel("VIDEO A.xlsx")
```

```python
def clean_dataframe(df):
    # Drop unnamed columns
    df = df.loc[:, ~df.columns.str.contains('Unnamed', case=False)]

    # Remove non-numeric rows
    df = df[pd.to_numeric(df['Video position (%)'], errors='coerce').notna()]

    # Convert columns to numeric
    df['Video position (%)'] = pd.to_numeric(df['Video position (%)'])
    df['Absolute audience retention (%)'] = pd.to_numeric(df['Absolute audience retention (%)'], errors='coerce')

    # Reset index
    df = df.reset_index(drop=True)

    return df


df_A = clean_dataframe(video_A)

```

```python
df_A
```

```python
missing_mask = df_A['Absolute audience retention (%)'].isna()

# Predict missing values
df_A.loc[missing_mask, 'Absolute audience retention (%)'] = model.predict(df_A.loc[missing_mask, ['Video position (%)']])

# Save the updated dataset A
df_A.to_excel("A_filled.xlsx", index=False)

print("Missing values in A have been predicted and saved.")
```
